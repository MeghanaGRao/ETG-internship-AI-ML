#   Reading from Inventory

```python
import json
fd = open("records.json",'r')
r = fd.read()
fd.close()
records = json.loads(r)
sale = {}
flag = 0
```

# Adding New Item to The Inventory

```python
product_id = input("Enter product ID:")
product_name = input("Enter the Product Name:")
price = int(input("Enter the price:"))
quant = int(input("Enter quantity:"))
C_mark = input("Enter the color mark(Red or Green) of the product:")
Expdate = input("Enter expiry date of the product in YYYY-MM-DD format:")

records[product_id] = {"Product_Name": product_name, 'Price': price, 'Quant_Avl': quant, "C_Mark": C_mark, "ExpDate": Expdate}

js = json.dumps(records)

fd = open("records.json",'w')
fd.write(js)
fd.close()
```

# To Delete a product from the inventory

```python
product_id = input("Enter product ID to be deleted from the inventory:")
if product_id in records.keys():
    del records[product_id]
    print("The product has been deleted from the inventory")
else:
    print("There is no product with the product ID entered in the Inventory")

js = json.dumps(records)

fd = open("records.json",'w')
fd.write(js)
fd.close()
```

# Purchasing and updating the inventory

```python
ui_prod  = str(input("Enter the product_Id of the product: "))
ui_quant = int(input("Enter the quantity you want to buy: "))
if (records[ui_prod]["Quant_Avl"] - ui_quant) >= 0 :
        flag += 1
        print("#############################################")
        print("Product:", records[ui_prod]['Product_Name'])
        print("Price:", records[ui_prod]['Price'])
        print("Quantities you're buying:", ui_quant)
        if records[ui_prod]["C_Mark"] == "Green":
            print("This is product is vegetarian")
        else:
            print("The color mark is Non-vegetarian")
        print("Use the product by",records[ui_prod]["ExpDate"])
        print("Billing Amount: ", records[ui_prod]['Price'] * ui_quant)
        print("##############################################")
        records[ui_prod]['Quant_Avl'] = records[ui_prod]['Quant_Avl'] - ui_quant
        sale[flag] = {"Product_Name": records[ui_prod]["Product_Name"], "Price": records[ui_prod]['Price'], "Quantities_Sold":ui_quant, "C_Mark":records[ui_prod]["C_Mark"], "Amount_Earned":records[ui_prod]['Price'] * ui_quant}
        
else:
        print("There are not enough Quantities Available")
        print("The number of quantites Available are:",records[ui_prod]["Quant_Avl"])
        print("Would you like to buy those?(Yes or No)")
        if input() == "Yes":
            flag += 1
            print("#############################################")
            print("Product:", records[ui_prod]['Product_Name'])
            print("Price:", records[ui_prod]['Price']) 
            print("Quantities you're buying:",records[ui_prod]["Quant_Avl"])
            if records[ui_prod]["C_Mark"] == "Green":
                print("This is product is vegetarian")
            else:
                print("The color mark is Non-vegetarian")
            print("Use this product by",records[ui_prod]["ExpDate"])
            print("Billing Amount: ", records[ui_prod]['Price'] * records[ui_prod]["Quant_Avl"])
            print("Have a Good day!")
            print("#############################################")
            sale[flag] = {"Product_Name": records[ui_prod]["Product_Name"], "Price": records[ui_prod]['Price'], "Quantities_Sold":records[ui_prod]["Quant_Avl"], "C_Mark":records[ui_prod]["C_Mark"], "Amount_Earned":records[ui_prod]['Price'] * records[ui_prod]["Quant_Avl"]}
            records[ui_prod]['Quant_Avl'] = 0
        else:
            print("Thanks for your patience, see you soon!")
# Delete the product from the inventory if the quantities available is zero after purchase       
if records[ui_prod]["Quant_Avl"] == 0:
        del records[ui_prod]
# Update the Inventory after purchase
js = json.dumps(records)
fd = open("records.json",'w')
fd.write(js)
fd.close()
```
